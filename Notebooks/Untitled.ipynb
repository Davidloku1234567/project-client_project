```python
import pandas as pd
import numpy as np
import multiprocessing
multiprocessing.cpu_count()
from os import listdir
from os.path import isfile, join
```

```python
onlyfiles = [f for f in listdir('../CSVs/weather_data/') if isfile(join('../CSVs/weather_data/', f))]
```

```python
def lowering(df):
    lowered_columns = []
    columns = df.columns
    for item in columns:
        lowered_columns.append(item.lower())
    lowered_columns[0] = 'county' #fixes the county index issue.
    df.columns = lowered_columns
    return df
```

```python
county_list = []
#for j in range(len(onlyfiles)):
for i in onlyfiles:
    path_csv = "../CSVs/weather_data/"+i
    name_county = i.replace('.csv', '')
    exec("%s = pd.read_csv('%s')" % (name_county,path_csv))
    exec('lowering(%s)' % (name_county))
    county_list.append(name_county)
```
